```python
NumFilas = Inicial.loc[Inicial.departamento == "LAMBAYEQUE","p2_1_02_006_2_1"].shape[0]
Inicial.loc[Inicial.departamento == "LAMBAYEQUE","p2_1_02_006_2_1"].value_counts().sort_index()*100/NumFilas
```

```python
# Calculemos el promedio de minutos de recreo para todos los departamentos
# donde tenemos datos

for dpto in Inicial.departamento.unique():
  # Para cada dpto mostremos el promedio
  print(" DEPARTAMENTO : %s -- Promedio_Minutos_Recreo : %.2f" %(dpto, Inicial.loc[Inicial.departamento == dpto,"p2_1_02_006_2_1"].mean()))
```

```python
type(Inicial.departamento.unique())
```

```python
# Construyamos un diccionario con la informacion calculada en la celda anterior
# Metodo fromkeys
DptosRecreo = {}.fromkeys(Inicial.departamento.unique())
DptosRecreo
for dpto in DptosRecreo.keys():
  DptosRecreo[dpto] = Inicial.loc[Inicial.departamento == dpto,"p2_1_02_006_2_1"].mean()

DptosRecreo
```

```python
# Transformemos este objeto DptosRecreo(DICCIONARIO) en un objeto de pandas
DptosRecreo = pd.Series(DptosRecreo)
DptosRecreo
```

```python
DptosRecreo.rename_axis("Columna").reset_index(name="NumFaltantes")
```

```python
DptosRecreo.shape
```

```python
DptosRecreo.to_csv("PromedioTiempoRecreo_Dpto.csv")
```

```python
# Calculo de la moda para un departamento
Inicial.loc[Inicial.departamento == "LAMBAYEQUE","p2_1_02_006_2_1"].mode()[0]
```

```python
# Calculo del numero de colegios para un departamento
Inicial.loc[Inicial.departamento == "LAMBAYEQUE","p2_1_02_006_2_1"].shape[0]
```

```python
# PAra pandas :
  # Un objeto unidimensional : pd.Series
  # Un objeto bidimensional : pd.DataFrame
```

```python
# Calculemos la moda y el numero de colegios para todos los departamentos
# Almacenando esta informacion en listas
ListaModa = list()
ListaNumColegios = list()
for dpto in Inicial.departamento.unique():
  ListaModa.append(Inicial.loc[Inicial.departamento == dpto,"p2_1_02_006_2_1"].mode()[0])
  ListaNumColegios.append(Inicial.loc[Inicial.departamento == dpto,"p2_1_02_006_2_1"].shape[0])
```

```python
# Agregamos columnas a DptosRecreo
DptosRecreo = pd.DataFrame(DptosRecreo)
#relacion [columna]=lista
DptosRecreo["Moda"] = ListaModa
DptosRecreo["NumColegios"] = ListaNumColegios
```
