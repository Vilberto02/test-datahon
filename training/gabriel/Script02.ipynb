```python
# Valores faltantes
HouseFaltantes = df.isna().sum().rename_axis("Columna").reset_index(name = "NumFaltantes").sort_values(by = "NumFaltantes", ascending = False)
HouseFaltantes
```

```python
# Agreguemos la columna "Porcentajes"
HouseFaltantes["Porcentajes"] = HouseFaltantes["NumFaltantes"]*100/df.shape[0]
HouseFaltantes
```

```python
HouseFaltantes.shape
```

```python
# Politica/Estrategia para la eliminacion de columnas que no aporten informacion
  # Columnas con mas del 40% de valores faltantes : SE ELIMINAN

# Primer paso : Construccion de una lista con los nombres de columnas a eliminar
ListaColumnasEliminar = HouseFaltantes.loc[HouseFaltantes.Porcentajes >= 40, "Columna"].to_list()
ListaColumnasEliminar


# Segundo Paso : Removamos las columnas que estan en ListaColumnasEliminar (metodo drop)
df2 = df.drop(ListaColumnasEliminar, axis = 1)
```

```python
df2.isna().sum().rename_axis("Columna").reset_index(name = "NumFaltantes").sort_values(by = "NumFaltantes", ascending = False)
```

isnull(): Es un método de pandas que se usa para detectar valores nulos o faltantes (NaN o None) en un DataFrame o Series.

isna(): Es una versión alternativa del mismo método. En realidad, isna() es un alias de isnull(), lo que significa que realizan la misma tarea y tienen el mismo comportamiento.

# Que hacemos con las columnas que aun tienen valores faltantes

```python
HouseFaltantes2 = df2.isnull().sum().rename_axis("Columna").reset_index(name = "Num.Faltantes").sort_values(ascending=False, by = "Num.Faltantes")
HouseFaltantes2["Porcentajes"] = HouseFaltantes2["Num.Faltantes"]*100/df2.shape[0]
ListaColumnasFaltantes2 = HouseFaltantes2.loc[HouseFaltantes2.Porcentajes >0 , "Columna"].to_list()
ListaColumnasFaltantes2
```

```python
df2[ListaColumnasFaltantes2].dtypes
```

```python
# Hagamos una lista con las variables (con valores faltantes) de tipo punto flotante
# A esas columnas, rellenaremos sus valores faltantes con  media
ListaValNum = []
for col in ListaColumnasFaltantes2:
  if df2[col].dtypes == "float64":
    ListaValNum.append(col)
ListaValNum
```

```python
# Construyamos una lista con las variables de tipo object
ListaValCuali = list()
for col in ListaColumnasFaltantes2:
  if df2[col].dtypes == "object":
    ListaValCuali.append(col)
ListaValCuali
```

PYTHON NO TOMA VALORES NULOS EN PROCESO DE SUMA PROMEDIO ETC


 Python, usando pandas, fillna() es un método que se utiliza para reemplazar valores nulos (NaN) en un DataFrame o Series con un valor específico

```python
# Rellenemos las columnas de tipo numericas con la media
for col in ListaValNum:
  df2[col] = df2[col].fillna(value = df2[col].mean())
```

```python
# Rellenemos las columnas cualitativas con la moda
for col in ListaValCuali:
  df2[col] = df2[col].fillna(value = df2[col].mode()[0])
```

```python
# Verifiquemos que el DataFrame df2 no tenga valores faltantes
df2.isnull()
```
